# OPL Data Exploration #

The purpose of this notebook is to practice using and to explore the OPL bulk dataset 

```python
import pandas as pd
import requests
import zipfile
import io
import re

# OPL dataset location
url_OPL = "https://openpowerlifting.gitlab.io/opl-csv/files/openpowerlifting-latest.zip"

# GET request
response_OPL = requests.get(url_OPL)

# unzip if successful
if response_OPL.status_code == 200:
    with zipfile.ZipFile(io.BytesIO(response_OPL.content)) as z:
        # get all file names in zip
        file_list = z.namelist()

        # define expected filename
        file_name_pattern = re.compile(r"openpowerlifting-\d{4}-\d{2}-\d{2}\/openpowerlifting-\d{4}-\d{2}-\d{2}-.{8}\.csv")

        # find first matching file
        csv_OPL = None
        for file in file_list:
            if file_name_pattern.match(file):
                csv_OPL = file
                break

        if csv_OPL == None:
            print("Could not find desired csv file in returned zip file.")
        
        # open file and import to df
        with z.open(csv_OPL) as f:
            df = pd.read_csv(f)
    
    print("CSV file downloaded and imported to dataframe successfully.")
    print("DataFrame shape: ", df.shape)
else:
    print("Failed to download OPL csv, response code: ", response_OPL.status_code)
```
